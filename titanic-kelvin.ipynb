```python
# Random Froest on Sept 13 2021

from sklearn.ensemble import RandomForestClassifier
```

```python
y = train["Survived"]
# In R, y <- train$Surivied

features = ["Pclass", "Sex", "SibSp", "Parch"]

X_train = pd.get_dummies(train[features])
X_test = pd.get_dummies(test[features])

model = RandomForestClassifier(n_estimators = 150, max_depth = 5, random_state = 42) #Defining the machine learning model 

model.fit(X_train, y)

predictions = model.predict(X_test)

output = pd.DataFrame({'PassengerId': test.PassengerId, 'Survived': predictions}) # Generate a Dataframe with two columns, "PassengerId" and "Survived"
output.to_csv('kelvin_submission.csv', index=False)
```
