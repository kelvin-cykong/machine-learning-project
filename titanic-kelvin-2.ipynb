```python
# Dropping columns that not useful

train = train.drop(['Name', 'Age', 'Ticket', 'Fare', 'Cabin', 'Embarked'], axis=1)

test = test.drop(['Name', 'Age', 'Ticket', 'Fare', 'Cabin', 'Embarked'], axis=1)
```

```python
# Encode Sex Column
# Importing LabelEncoder

from sklearn.preprocessing import LabelEncoder
label_encoder_sex = LabelEncoder()

train.iloc[:,3] = label_encoder_sex.fit_transform(train.iloc[:,3])
test.iloc[:,2] = label_encoder_sex.fit_transform(test.iloc[:,2])
```

```python
# Rearrange Columns
train = train[['PassengerId', 'Pclass', 'Sex', 'SibSp', 'Parch', 'Survived']]
test = test[['PassengerId', 'Pclass', 'Sex', 'SibSp', 'Parch']]
```

```python
X_train = train.iloc[:, 0:4] # Input
y_train = train.iloc[:, 5] # Train Output
```

```python
# Normalize data
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.fit_trainsform(test)
```

```python
# Import Keras for model building
import keras

from keras.models import Sequential
from keras.layers import Dense

model = Sequential()

model.add(Dense(3, activation = 'relu', input_dim = 4))
model.add(Dense(2, activation = 'relu'))
model.add(Dense(1, activation = 'sigmoid'))
```

```python
model.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])
model.fit(X_train, y_train, epochs = 50)
```
